```python
import numpy as np
import pandas as pd
```

```python
# read in data for account 1

xlsx_file = pd.ExcelFile('output_acct1.xlsx')
df = xlsx_file.parse('1', header=0)
```

```python
df.head()
```

```python
# create datetime column

datetime = pd.date_range('1/1/2017 00:15', '1/1/2018 00:00', freq='15min')
```

```python
df['DateTime'] = datetime
df.head()
```

```python
# create weekday column
# 1/1/2017 is Sunday

WeekDay_list = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
WeekDay = []

for x in range(35040):
    WeekDay.append(WeekDay_list[(x//96)%7])
    
len(WeekDay)
```

```python
df['WeekDay'] = WeekDay
df.head(97)
```

```python
# create season column

Season = []


for i in range(len(df)):
    if df['DateTime'][i].month in range(5,11):
        Season.append('Summer')
    else:
        Season.append('Winter')
  
len(Season)
```

```python
df['Season'] = Season
df.head()
```

```python
# create tier column

Tier = []
Weekends = ['Saturday', 'Sunday']


for i in range(len(df)):
    if (df['Season'][i]=='Summer') and (df['WeekDay'][i] not in Weekends):
        if df['DateTime'][i].hour in range(11,18):
            Tier.append('OnPeak')
        elif df['DateTime'][i].hour in range(6,11) or df['DateTime'][i].hour in range(18,22):
            Tier.append('SemiPeak')
        else:
            Tier.append('OffPeak')
    elif (df['Season'][i]=='Winter') and (df['WeekDay'][i] not in Weekends):
        if df['DateTime'][i].hour in range(17,20):
            Tier.append('OnPeak')
        elif df['DateTime'][i].hour in range(6,17) or df['DateTime'][i].hour in range(20,22):
            Tier.append('SemiPeak')
        else:
            Tier.append('OffPeak')
    else:
        Tier.append('OffPeak')
```

```python
df['Tier'] = Tier
df.head()
```

```python
# switch column order

cols = df.columns.tolist()
cols
```

```python
cols = cols[-4:] + cols[:-4]
cols
```

```python
df = df[cols]
df.head()
```

```python
# add bill rate column

rate = []

for i in range(len(df)):
    if df['Season'][i] == 'Summer':
        if df['Tier'][i] == 'OnPeak':
            rate.append(0.11715)
        elif df['Tier'][i] == 'SemiPeak':
            rate.append(0.10752)
        else:
            rate.append(0.07696)
    else:
        if df['Tier'][i] == 'OnPeak':
            rate.append(0.10543)
        elif df['Tier'][i] == 'SemiPeak':
            rate.append(0.09000)
        else:
            rate.append(0.06875)
```

```python
df['Rate'] = rate
df.head()
```

```python
# calculate electricity bill

df['Bill'] = df['Net Usage (kWh)'] * df['Rate']
df.head()
```

```python
# write the results into excel file

writer = pd.ExcelWriter('electric_bill.xlsx')
df.to_excel(writer)
writer.save()
```
